### Question 13

Perform the following tasks:

- Another useful plot you can use to look at correlations between features both numerically and visually is the correlation heatmap.
- The first part of the code below plots a diagonal matrix of correlations between features. The second part produces the heat map, where the colour key shows the value of the correlations between features.
<code>
# Calculate the Pearson Correlation Coefficient between all features
plt.figure(figsize=(10,8))
corr_df = tips.corr(method="pearson") 
display(corr_df)
sns.heatmap(corr_df, cmap=sns.diverging_palette(180, 20, as_cmap=True),xticklabels=corr_df.columns.values,
                                                                       yticklabels=corr_df.columns.values)
</code>

- What can you conclude has a stronger correlation? Answer this question in the **Conclusions** cell below.

```python
plt.figure(figsize=(10,8))
numeric_columns = tips.select_dtypes(include=['number'])
corr_df = numeric_columns.corr(method="pearson")
display(corr_df)

sns.heatmap(corr_df, cmap=sns.diverging_palette(180, 20, as_cmap=True),xticklabels=corr_df.columns.values,
                                                                       yticklabels=corr_df.columns.values)
```

**Conclusions**: 

- Tip and total bill have the strongest correlation.
